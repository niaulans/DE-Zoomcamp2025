```python
df1.lpep_pickup_datetime = pd.to_datetime(df1.lpep_pickup_datetime)
df1.lpep_dropoff_datetime = pd.to_datetime(df1.lpep_dropoff_datetime)
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
print(pd.io.sql.get_schema(df1, name='green_taxi_trips', con=engine))
```

```python
print(pd.io.sql.get_schema(df2, name='taxi_zone', con=engine))
```

```python
df1_iter = pd.read_csv('green_tripdata_2019-10.csv.gz', iterator=True, chunksize=100000)
```

```python
while True:
    try:
        t_start = time()
        df = next(df1_iter)

        df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
        df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
        
        df.to_sql(name='green_taxi_trips', con=engine, if_exists='append')

        t_end = time()
        print('Inserted another chunk, took %.3f second' % (t_end - t_start))
        
    except StopIteration:
        print("All chunks have been inserted. Process completed!")
        break
```

```python
engine.connect()
```

```python
query = """
SELECT 
    COUNT(*)
FROM 
    green_taxi_trips;
"""

pd.read_sql(query, con=engine)
```
